```python
import pandas as pd
df = pd.read_csv("college_event_feedback_dataset.csv")
df.head()
```

```python
df['Feedback Text'] = df['Feedback Text'].str.lower().str.strip()
```

```python
from textblob import TextBlob
df['Polarity'] = df['Feedback Text'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Rating Distribution
sns.histplot(df['Rating'], bins=5)
plt.title("Rating Distribution")
plt.show()

# Sentiment by Department
sns.barplot(x='Department', y='Polarity', data=df)
plt.title("Sentiment by Department")
plt.show()
```

```python
from wordcloud import WordCloud
text = " ".join(df['Feedback Text'])
wordcloud = WordCloud(background_color="white").generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
print("Average Rating:", df['Rating'].mean())
print("Average Sentiment:", df['Polarity'].mean())
```
